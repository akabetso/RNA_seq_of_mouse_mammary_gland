```python
import pandas as pd
import io
```

```python
# Load the featureCounts output into a DataFrame
counts_file = "../results/featureCounts/cleaned_fcounts.txt" 
counts_df = pd.read_csv(counts_file, sep='\t')
counts_df.head()
```

```python
experiment_file = "../data/targets2.txt" 
experiment_file_df = pd.read_csv(experiment_file, sep='\t')
experiment_file_df.head()
```

```python
# Create a dictionary to map file names to sample names
sample_mapping = dict(zip(experiment_file_df.iloc[:, 4].str.replace('.fastq.gz', '_trimmed_aligned.bam'), experiment_file_df.iloc[:, 1]))
sample_mapping
# Rename the columns in the counts DataFrame
counts_df.rename(columns=sample_mapping, inplace=True)
counts_df
#counts_df = counts_df.drop(columns=['Unnamed: 13'])
counts_df
# Save the updated DataFrame to a new file
counts_df.to_csv("../results/featureCounts/counts_matrix.csv", index=False)
```
